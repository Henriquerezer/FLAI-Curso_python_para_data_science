**Padrão da documentação**

criar 10 bins


Fonte : https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

```python
plt.figure(figsize=(15,5))
plt.hist(data.age, bins = np.arange(0,101,1), rwidth = 0.9 );
# o comandao np.arange significa =  que estou definindo uma lista de números que começa em 0, termina em 101, indo de 1 em 1
# o comando rwidth significa = Largura da caixa da direita ( largura da caixa ao lado)
```

```python
#Colocando legenda

plt.figure(figsize=(15,5))
plt.hist(data.age, bins = np.arange(0,101,1), rwidth = 0.9 )

#Tamanho da legenda
plt.xlabel('Idade', fontsize = 16)
plt.ylabel('Contagem de pessoas', fontsize = 16)

#tamanho do eixo
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16);

#Colocando grade no fundo
plt.grid()
```

```python
#Aplicando argumento density, se o argumento estiver em True. A coluna está em formato de probabilidade

#Colocando legenda

plt.figure(figsize=(15,5))
plt.hist(data.age, bins = np.arange(0,101,1), rwidth = 0.9, density = True)

#Tamanho da legenda
plt.xlabel('Idade', fontsize = 16)
plt.ylabel('Probabilidade por idade', fontsize = 16)

#tamanho do eixo
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16);

#Colocando grade no fundo
plt.grid()

```

```python
plt.figure(figsize=(15,5))
plt.hist(data.age, bins = np.arange(0,101,1), rwidth = 0.9, density = True)

#linha vertical pararela ao meu eixo
plt.axvline(29, color = 'black')
#linha horizontal ao meu eixo
plt.axhline(0.0125, color = 'k')
plt.xlabel('Idade', fontsize = 16)
plt.ylabel('Probabilidade por idade', fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.grid()
```

```python
plt.figure(figsize=(15,5))
plt.hist(data.age, bins = np.arange(0,101,1), rwidth = 0.9, density = True)

plt.axvline(29, color = 'black')
plt.axhline(0.0125, color = 'k')

plt.xlabel('Idade', fontsize = 16)
plt.ylabel('Probabilidade por idade', fontsize = 16)

#posso passar para para o meu histograma aonde eu quero esses ticks, passando uma lista

plt.xticks(fontsize = 16)
plt.yticks([0,0.005,0.010,0.0125,0.015,0.020,0.025],fontsize = 16)
plt.grid()
```

**Distribuição de variável conjunta**

Parâmetros sem um ' = ' na frente, são parâmetros obrigatórios

Fonte : https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist2d.html

```python
plt.figure(figsize=(15,5))
#plt.hist(data.age, bins = np.arange(0,101,1), rwidth = 0.9, density = True)
plt.hist2d(data.age, data.MonthlyIncome, bins = 30, density = True)

plt.xlabel('Idade', fontsize = 16)
plt.ylabel('Probabilidade por idade', fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.grid()
```

```python
plt.figure(figsize=(15,5))
#plt.hist(data.age, bins = np.arange(0,101,1), rwidth = 0.9, density = True)
plt.hist2d(data.age, data.MonthlyIncome, bins = 30, density = True)

plt.xlabel('Idade', fontsize = 16)
plt.ylabel('Probabilidade por idade', fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16);
```

```python
plt.figure(figsize=(15,5))
#plt.hist(data.age, bins = np.arange(0,101,1), rwidth = 0.9, density = True)
plt.hist2d(data.age, data.MonthlyIncome, bins = 30, density = True)
plt.colorbar()

plt.xlabel('Idade', fontsize = 16)
plt.ylabel('Probabilidade por idade', fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16);
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.hist(data.MonthlyIncome, bins = 30, density =True, orientation = 'horizontal', rwidth = 0.9)
plt.xticks(fontsize = 16, rotation = 45)
plt.yticks(fontsize = 16)

plt.subplot(1,2,2)
plt.hist2d(data.age, data.MonthlyIncome, bins = 30, density = True)
plt.colorbar()

plt.xlabel('Idade', fontsize = 16)
plt.ylabel('Probabilidade por idade', fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16);
```

```python
plt.figure(figsize=(15,5))

plt.subplot(2,2,1)
plt.hist(data.MonthlyIncome, bins = 30, density =True, orientation = 'horizontal', rwidth = 0.9)
plt.xticks(fontsize = 16, rotation = 45)
plt.yticks(fontsize = 16)

plt.subplot(2,2,2)
plt.hist2d(data.age, data.MonthlyIncome, bins = 30, density = True)
plt.colorbar()

plt.subplot(2,2,4)
plt.hist(data.age, bins = 30, density =True, orientation = 'vertical', rwidth = 0.9)
plt.xticks(fontsize = 16, rotation = 45)
plt.yticks(fontsize = 16)

plt.xlabel('Idade', fontsize = 16)
plt.ylabel('Probabilidade por idade', fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16);
```
